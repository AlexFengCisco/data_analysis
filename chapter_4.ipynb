### Pandas Data Aggregation

```python
from numpy.random import seed
from numpy.random import rand
from numpy.random import random_integers
import sys
```

```python
seed(42)

df = pd.DataFrame({'Weather' : ['cold', 'hot', 'cold', 'hot',
   'cold', 'hot', 'cold'],
   'Food' : ['soup', 'soup', 'icecream', 'chocolate',
   'icecream', 'icecream', 'soup'],
   'Price' : 10 * rand(7), 'Number' : random_integers(1, 9, size=(7,))})
```

```python
df
```

```python
weather_group = df.groupby("Weather")  
```

```python
weather_group
```

```python
for item in weather_group:
    print(item)
```

```python
weather_group.mean().Price.iloc[[0]] # seems pandas can ignore non computable data ,leave int float etc
```

```python
dir(weather_group.mean().Price)
```

```python
wf_group = df.groupby(["Weather","Food"])
```

```python
for item in wf_group:
    print(item)
```

```python
print(wf_group.agg([np.mean,np.median,np.max,np.min]))
```
